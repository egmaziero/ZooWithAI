```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
import seaborn as sns
```

## Passo 1: Leitura dos dados
Leitura dos bancos de dados em DataFrame Pandas. 

```python
caminho_dados_lazaro = 'Datasets/Lazaro/dados_pre_processados_lazaro.csv'

df = pd.read_csv(caminho_dados_lazaro, index_col='Date', parse_dates=True)

# Remove as linhas com o sensor 434138
df = df[df.SensorID != 434138]
```

### Gráficos de dispersão

```python
fig = plt.figure()
ax = fig.add_subplot(111)

for s in df.Comportamento.unique():
    ax.scatter(df.X[df.Comportamento==s],df.Y[df.Comportamento==s],label=s)
    
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

for s in df.Comportamento.unique():
    ax.scatter(df.X[df.Comportamento==s],df.Z[df.Comportamento==s],label=s)
    
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for s in df.Comportamento.unique():
    ax.scatter(df.X[df.Comportamento==s],df.Y[df.Comportamento==s],df.Z[df.Comportamento==s],label=s)
    
ax.legend()
```
