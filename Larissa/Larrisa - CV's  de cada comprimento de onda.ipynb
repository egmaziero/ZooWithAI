```python
cols = [i for i in range(23, 2226)]
```

```python
# Leitura em DataFrame com colunas específicas
df =  pd.read_excel('Datasets/Larissa/Matriz completa - Larissa MD04.xlsx', usecols=cols)
```

```python
df
```

```python
#define function to calculate cv
cv = lambda x: np.std(x) / np.mean(x) * 100 
```

```python
df.apply(cv)
```

```python
df
```

```python
cof_variacao = df.apply(cv)
```

```python
cof_variacao = pd.DataFrame(cof_variacao)
```

```python
cof_variacao
```

```python
cof_variacao = cof_variacao.reset_index()
```

```python
cof_variacao.rename({'index':'Comprimento de onda', 0:'Coeficiente de variação'}, axis=1, inplace=True)
```

```python
cof_variacao.to_excel('larissa_coef_variacao_comprimento_onda.xlsx', index=False)
```
