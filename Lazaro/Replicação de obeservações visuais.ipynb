```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Datasets/Lazaro/Dados observações visuais/dados_observacoes_lazaro.csv")
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df = df.set_index(['Date'])
```

```python
df
```

```python
df.Animal.value_counts()
```

```python
animais = df.Animal.value_counts().keys().to_list()
```

```python
for animal in animais:
    df_temp = df[df['Animal'] == animal]
    new_df = pd.DataFrame(dtype='datetime64[ns]', columns=df_temp.columns)
    lista_de_datas = np.unique(df_temp.index.date)
    for data in lista_de_datas:
        temp = pd.DataFrame(pd.date_range(start=df_temp.loc[data.isoformat()].index[0], end=df_temp.loc[data.isoformat()].index[-1], freq='5min'))
        new_df = pd.concat([new_df, temp])
    new_df = new_df.set_index([0])
    new_df['Animal'] = animal
    i = 0
    for date in new_df.index:
        found_interval = False
        while(not(found_interval)):
            if(date >= df_temp.index[i] and date < df_temp.index[i+1]):
                new_df.loc[date] = df_temp.iloc[i]
                found_interval = True
            else:
                i +=1
    new_df.to_csv('animal_' + str(animal) + '.csv')
```

```python
f = open("resultados5em5.txt", "a")
f.write("5 em 5 minutos\n\n")
for animal in animais:
    df_temp = pd.read_csv('animal_' + str(animal) + '.csv')
    
    f.write(f"Animal {animal}:\n\n" +
            str(df_temp.Comportamento.value_counts(normalize=True)*100) +'\n\n')
f.close()
```

```python
f = open("resultadosSegundoemSegundo.txt", "a")
f.write("Segundo em segundo\n\n")
for animal in animais:
    df_temp = pd.read_csv('animal_' + str(animal) + '.csv')
    
    f.write(f"Animal {animal}:\n\n" +
            str(df_temp.Comportamento.value_counts(normalize=True)*100) +'\n\n')
f.close()
```
