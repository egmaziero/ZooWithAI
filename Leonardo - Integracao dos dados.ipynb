## 2. Integração das observações visuais e dados de sensor

```python
df_sensor = pd.read_csv('Datasets/Leonardo/dados_accelerometer_sensor_leonardo.csv', sep=',')
```

```python
df_sensor
```

```python
df_sensor.dtypes
```

```python
df_sensor['Date'] = pd.to_datetime(df_sensor['Date'])
```

```python
df_sensor['Date'] = df_sensor['Date'].astype(str)
```

```python
df_observacoes = pd.read_csv('Datasets/Leonardo/dados_comportamentais_Leonardo.csv', sep=',')
```

```python
df_observacoes
```

```python
df_observacoes.dtypes
```

```python
df_observacoes['Date'] = df_observacoes['Date'] + " " + df_observacoes['Time']
```

```python
df_observacoes.drop('Time',axis=1, inplace=True)
```

```python
df_observacoes['Date'] = pd.to_datetime(df_observacoes['Date'])
```

```python
df_observacoes['Date'] = df_observacoes['Date'].astype(str)
```

```python
# Dicionário de observação de cada sensor
dic_sensor = df_observacoes['SensorID'].value_counts()
```

```python
dic_sensor
```

```python
# Dicionário com o dataframe de cada animal
dic_df_sensor = {}

# Cria uma dataframe para cada animal
for i in dic_sensor.keys():
    value = df_observacoes.loc[df_observacoes['SensorID'] == i, ['Date', 'Comportamento', 'Animal']].reset_index(drop=True)
    dic_df_sensor[i] = value
```

```python
dic_df_sensor[175810]
```

```python
# Inicializa o DataFrame de união com valores indefinidos
df = df_sensor
df['Comportamento'] = 'Indefinido'
```

```python
df
```

```python
print(df.at[0,'Date'][11:16])
```

```python
# Algorítmo que faz a integração dos dados de sensor com as observações visuais
ultimo_k = 0
maior_que_o_intervalo = False
qtd_dados = df_sensor.shape[0]

for i in dic_sensor.keys(): # i controla o animal
    print(i)
    ultimo_k = 0
    k = 0                             # k itera os dados de sensor
    for j in range(dic_sensor[i] - 1): # j controla a observação visual do animal i
        maior_que_o_intervalo = False
        k = ultimo_k
        while(k < qtd_dados and (not(maior_que_o_intervalo))): 
            
            antes_das_seis = df.at[k,'Date'][11:16] < '18:00'
            
            if ((df.at[k,'SensorID'] == i) and df.at[k,'Date'] >= dic_df_sensor[i].at[j,'Date'] and antes_das_seis):
                if (df.at[k,'Date'] < dic_df_sensor[i].at[j + 1,'Date']):
                    df.at[k,'Comportamento'] = dic_df_sensor[i].at[j,'Comportamento']
                    ultimo_k = k + 1
                elif((j + 1) == (dic_sensor[i] - 1) and df.at[k,'Date'][11:13] < '18'): # condição de última linha
                    df.at[k,'Comportamento'] = dic_df_sensor[i].at[j + 1,'Comportamento']
                else:
                    maior_que_o_intervalo = True
                    ultimo_k = k
            k += 1
```

```python
df
```

```python
df = df[df['Comportamento'] != 'Indefinido']
```

```python
df
```

```python
df.Comportamento = df.Comportamento.replace(to_replace=[1, 2, 3, 4, 5, 6, 7], value=['Pastejando', 'Ruminando EM PE', 'Ruminando DEITADO', 'Ocio EM PE', 'Ocio DEITADO', 'Agua', 'Locomovendo'])
```
