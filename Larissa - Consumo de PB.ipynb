# 1. Leitura dos dados

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
def imprimirAnaliseGrafica(y_test, preds, model):
    # Plot de dispersão
    plt.figure(figsize=(12,6))
    plt.title("Dipersão dos dados reais vs predições usando " + type(model).__name__)
    sns.scatterplot(x=y_test.index, y=y_test, label="Real")
    sns.scatterplot(x=y_test.index, y=preds, label="Predito")
    plt.figure(figsize=(12,6))
    plt.title("Real vs Predito usando " + type(model).__name__)
    # Plot de regressão
    sns.regplot(x=y_test, y=preds)
    plt.xlabel("Real")
    plt.ylabel("Predito")
    plt.legend(["Ponto","R² = " + str("%.2f"%r2_score(y_test, preds))])
```

```python
# Leitura dos dados em uma variável DataFrame
df =  pd.read_excel('Datasets/Larissa/Consumo de Proteina - Larissa MD04.xlsx')
```

```python
df
```

```python
y = df['Consumo de PB']
X = df.iloc[:, 5:]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

# 2. Testes

## Random Forrest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state=0)
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_preds = rf_model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, rf_preds))
```

```python
imprimirAnaliseGrafica(y_test, rf_preds, rf_model)
```
